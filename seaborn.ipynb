```python
# introduction to seaborn

# import seaborn
import seaborn as sns

# Apply the deafault theme

sns.set_theme()


tips = sns.load_dataset("tips")
tips.sample(3)

# create a visualization
sns.relplot(
    data = tips,
    x = 'total_bill', y = 'tip', col = 'time',
    hue = 'smoker', style = 'smoker', size = 'size'
)
```

```python
dots = sns.load_dataset('dots')
sns.relplot(
    data=dots, kind='line', # kind parameter let's you easily  switch to #line
    x='time', y='firing_rate', col='align',
    hue='choice', size='coherence', style='choice',
    facet_kws=dict(sharex=False)
)
```

```python
# statistical estimation
fmri = sns.load_dataset('fmri')
sns.relplot(
    data = fmri, kind='line',
    x='timepoint', y = 'signal', col='region',
    hue='event', style='event'
);
fmri

```

```python
# statistical estimation in seaborn goes beyond descriptive statistics. enhance a scatterplot by including a linear regression model
sns.lmplot(
    data=tips,
    x='total_bill', y='tip', col='time', hue='smoker'
           
)
```

```python
# Distribution representation
# sns.displot(data=tips, x='total_bill', col="time", kde=True)

# seaborn promotes techniques that are powerful but less familiar such as calculating and plotting the empirical cumulative distribution function of the data.
sns.displot(data=tips, kind='ecdf', x='total_bill', col='time', hue='smoker', rug=True)
```

```python
# plots for categorical data
# they can be accessed through catplot(). this plots offer different level of granularity. 
# you may wish to see every obervation by drawing a swarm plot, a scatter plot that adjusts the positions of the point along thecategorical axis so that they don't overlap
sns.catplot(data=tips, kind='swarm', x='day', y='total_bill', hue='smoker')
```

```python
# alternatively you can use the kernel density estimation to represent the underlying distribution that the point are sampled from:
sns.catplot(data=tips, kind='violin', x='day', y='total_bill', hue='smoker', split=True)
```

Relational, distributional and categorical



```python
penguins = sns.load_dataset('penguins')
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', multiple='stack')
```

```python
# Figure level functions interface with matplotlib through a seaborn object.
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
# kernel densitplot using the same code using kdeplot() select using the kind parameter
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack", kind='kde')
```

```python

# the most useful feature offered by the figure-level functions is that they can easily create figures with multiple subplots
sns.displot(data=penguins, x='flipper_length_mm', hue='species', col='species')
```
