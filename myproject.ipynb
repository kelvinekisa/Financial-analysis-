```python
# create a dataframe copy to be used in data cleaning
df1 = df.copy(deep=True)

# check for unique values
for column in df1:
    unique_values = df1[column].unique()
    print(f"Unique values in column, \n '{column}': {unique_values},\n")
```

```python
df1.columns
df1['yr_built'] = pd.to_datetime(df1['yr_built'], format='%Y')

# split into year, month and day
df1['Built_Year'] = df1['yr_built'].dt.year
df1['Built_Month'] = df1['yr_built'].dt.month
df1['Built_day'] = df1['yr_built'].dt.day

del df1['yr_built']

df1.sample(10)
```
